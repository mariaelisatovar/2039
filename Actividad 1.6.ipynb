```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('datasets/pokemon.csv')

df
```

```python
pd.pivot_table(df,
              index = 'Type 1',
              values = {'Attack', 'Defense'},
              aggfunc = 'mean')
```

```python
sns.boxplot(data = df.loc[df['Type 1'] == 'Grass',
                         ['Attack', 'Defense']]
           )
plt.title('box plot Grass')
plt.show()
```
