```python
df_total = df.dropna()
df_total.count()
```

```python

peso = df_total['peso']
cintura = df_total['cintura']
print(peso.count(),cintura.count())
```

```python
import statsmodels.api as sm
model = sm.OLS.from_formula('peso~cintura',  #peso = b0 + b1*cintura
                           data = df_total)
result1 = model.fit()
result1.summary()
```

```python
ft = result1.fittedvalues
ft
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

reg_plot= sns.regplot(x="cintura",
                     y="peso",
                     data=df_total,
                     color="b",
                     marker="+",
                     scatter_kws={"s":50})

reg_plot.set(xlabel="Magnitud de cintura",
      ylabel="Peso")
reg_plot.set_title("Regresión lineal de Peso con respecto a la cintura")
plt.show()
```
