```python
# import the packages we are going to be using

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import pandas as pd
#pd.set_option('display.max_columns', 100)
#import scipy.stats as stats

path = "antropometria-dataset.csv"

# Importar los datos
df = pd.read_csv(path)

df.head(10)
```

```python
# Guardar los datos de las variables de interÃ©s
df.fillna(method ='ffill', inplace = True) 
#revisar sustituciÃ³n 
df.head(10)
```

```python
#guardar los datos de las variables de interÃ©s
sistol = df['sistol']
diastol = df['diastol']
```

```python
#Se usa matplotlib.pyplot as plt
#Fijar el tamaÃ±o de las grÃ¡ficas en pulgadas 
plt.figure(figsize=(10,5))

# Visualizamos un histograma al lado del otro
#nÃºmero de renglones, nÃºmero de columnas, Ã­ndice
plt.subplot(1,2,1)
plt.hist(x = sistol, bins = 10)
plt.title("Sistol")

plt.subplot(1,2,2)
plt.hist(x = diastol, bins = 10)
plt.title("Diastol")

#busca los objetos grÃ¡ficos activos y los muestra 
plt.show()
```

```python
#guardar los datos de las variables de interÃ©s
peso = df['peso']
cintura = df['cintura']
```

```python
#Se usa matplotlib.pyplot as plt
#Fijar el tamaÃ±o de las grÃ¡ficas en pulgadas 
plt.figure(figsize=(10,5))

# Visualizamos un histograma al lado del otro
#nÃºmero de renglones, nÃºmero de columnas, Ã­ndice
plt.subplot(1,2,1)
plt.hist(x = peso, bins = 10)
plt.title("Peso")

plt.subplot(1,2,2)
plt.hist(x = cintura, bins = 10)
plt.title("Cintura")

#busca los objetos grÃ¡ficos activos y los muestra 
plt.show()
```

```python
# Generar un diagrama de dispersiÃ³n de peso vs. largo
plt.figure(figsize=(5,5))
plt.scatter(x = peso, y = cintura)
plt.xlabel("'peso'")
plt.ylabel("'cintura'")
plt.title("Diagrama de dispersiÃ³n de peso vs cintura")

plt.show()
```
