```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
ad = pd.read_csv('datasets/airline_data.csv',
                encoding = 'ISO-8859-1',
                low_memory=False)
ad.head(2)
```

```python
data = ad.sample(n = 500,
                random_state = 42)
```

```python
data.shape
```

```python
data.columns
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure(data = go.Scatter(x = data['Distance'],
                                 y = data['DepTime'],
                                 mode = 'markers',
                                 marker = dict(color = 'red')
                                 )
               )
fig.update_layout(title = 'Distance vs Departure time',
                 xaxis_title = 'Distance',
                 yaxis_title = 'Departure time')
fig.show()
```

```python
line_data = data.groupby('Month')['ArrDelay'].mean().reset_index()
line_data
```

```python
fig = go.Figure(data = go.Scatter(x = line_data['Month'],
                                 y = line_data['ArrDelay'],
                                 mode = 'lines',
                                 marker = dict(color = 'green')
                                 )
               )
fig.show()
```

```python
fig = px.bar(line_data,
            x = 'Month',
            y = 'ArrDelay',
            title = 'Bar chart')
fig.show()
```
